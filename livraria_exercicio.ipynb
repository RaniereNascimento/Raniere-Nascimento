```python
#  listar todos os generos de cada livro e por em ordem alfabetica
#  listar todos os subgeneros de cada livro e por em ordem alfabetica
#  listar todos os livros e colocar os preços em ordem numérica


livraria = {}

# Exemplos de livros já cadastrados

livraria['livro1'] = {'titulo': 'Norwegian Wood' ,
'gênero': 'Drama',
'subgênero': 'Romance',
'editora': 'Alfaguara' ,
'cópias' : '200',
'preço': '49,90'+' R$' }
 

livraria['livro2'] = {'titulo' :'1Q84 vol 1',
'gênero': 'Ficção' ,
'subgênero': 'Fantasia',
'editora': 'Alfaguara' ,
'cópias' : '130',
'preço': '45,90'+' R$' }


livraria['livro3'] = {'titulo' :'O Semhor dos anéis',
'gênero': 'Fantasia' ,
'subgênero': 'Aventura',
'editora': 'Martins-Fontes' ,
'cópias' : '130',
'preço': '65,00'+' R$' }



livraria['livro4'] = {'titulo' :'Harry Potter',
'gênero': 'Fantasia' ,
'subgênero': 'Aventura',
'editora': 'Rocco' ,
'cópias' : '160',
'preço': '47,00' +' R$'}


# Função para cadastrar novo livro
def Cadastrar():
  titulo = input("Digite o título")
  genero = input("Digite o genero")
  subgenero = input("Digite o genero")
  editora = input("Digite a editora")
  copias = input("Digite a quantidade de cópias")
  preco = input("Digite o preço")
  novo_livro = {'título' : titulo,
                'gênero': genero ,
                'subgênero': subgenero,
                'editora': editora , 
                'cópias' : copias, 
                'preço': preco + " R$" }

  novo_livro["titulo"] = titulo
  novo_livro["gênero"] = genero 
  novo_livro["subgênero"] = subgenero
  novo_livro["editora"] = editora
  novo_livro["cópias"] = copias
  novo_livro["preço"] = preco + " R$"           

# Função para consulta
def Consultar():
  for chave, valor in sorted(livraria.items()):
    gen = livraria.get(chave).get('gênero')
    sub = livraria.get(chave).get('subgênero') 
    tit = livraria.get(chave).get('titulo')
    prec = livraria.get(chave).get('preço')
    qtd  = livraria.get(chave).get('cópias')
    ed   = livraria.get(chave).get('editora')

    print("")
    print("**************************************************")
    print("--- Gênero: ---  ")
    print("")
    print('{}'.format(gen))
    print("")
    print("------Subgênero ------") 
    print("")
    print('{}'.format(sub))
    print("")
    print('-- LIVRO: {}  -editora: {}-  preço:: {}- '.format(tit,ed,prec))
    print('Estoque: {}'.format(qtd))
    print("**************************************************")
    print("")





# Menu
print("******************")
print("ESCOLHA UMA OPÇÃO:")
print("******************")
print("")
print("1.  CONSULTAR ACERVO GERAL")
print("2.  CADASTRAR NOVO TÍTULO")
print("")


opt = int(input("Digite o número da opção: "))

if opt ==1:
  Consultar()
  print(opt)
if  opt ==2:
  Cadastrar()

if opt !=1 and opt  !=2:
  print("Digite um número válido")
  

  




```
